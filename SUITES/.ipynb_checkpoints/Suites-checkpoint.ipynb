## II) Raisonnement par récurrence

### 1) Définition de "propriété mathématique"

Une propriété mathématique est un phrase mathématique (symbolisée ou en lettre) qui peut être qualifiée de vraie ou de fausse. 

> Exemple: Dire si ces propriétés vous semblent vraies ou fausses.
> 
> $P_1(n): 1+2+ ... + n = \frac{n(n+1)}{2}$  
> $P_2(n): U_n \leq 6 $  pour tout entier naturel $n$, sachant que $U_0 = 2$ et $U_{n+1}=U_n \times 2$.  
> $P_3(n): 5^{2n}-1$ est un multiple de 24 pour $n$ entier naturel.  

###  2) Principe de récurrence (axiome)

$P(n)$ désigne une propriété mathématique sur un entier naturel $n$.   
Pour démontrer que cette propriété est vraie quelque soit le rang $n$ (on dit: pour tout $n$ entier naturel), il faut procéder en 2 étapes.   
- Une des deux étapes est de prouver que si $P$ est vraie pour un rang  donné $n$, alors elle est vraie pour le rang suivant $n+1$, il s'agit de **l'hérédité** de cette propriété.  
- L'autre étape est triviale! Car pour que la propriété soit vraie pour tout $n$, faut-il qu'il y ait un premier rang pour lequel cette propriété soit vraie.  

<div class="alert alert-block alert-info">
    <b>Démontrons que $1+2+... + n = \frac{n(n+1)}{2}$ <u>pout tout n naturel</u>.</b><br> <br>
    <u> Initialisation:</u> <br>
        pour $ n = 1$, on a $\frac{1(1+1)}{2}=1.$  
        Donc $P(1)$ est vraie.<br><br>
    <u> Hérédité:</u> <br>
    On suppose que $P(n)$ est vraie pour un rang $n$ donné. On cherche à prouver que <u><b>pour un rang n donné</u></b> :<br>
      $ P(n)$ est vraie $\Longrightarrow P(n+1)$ est vraie <br><br>
    $ 1+2+...+n = \frac{n(n+1)}{2} \Longrightarrow  1+2+...+ (n+1) = \frac{(n+1)(n+2)}{2}$<br><br>
        Remplaçons les n premiers termes de notre somme par $P(n)$:
        $$ 1+2+...+n+(n+1) = \frac{n(n+1)}{2}+(n+1)$$<br>
        $$ 1+2+...+n+(n+1) =  \frac{n(n+1)}{2} + \frac{2(n+1)}{2} $$<br>
        $$ 1+2+...+n+(n+1) = \frac{n(n+1)+2(n+1)}{2}$$<br>
        $$ 1+2+...+n+(n+1) = \frac{(n+1)(n+2)}{2}$$<br>
        On en déduit que:
         $P(n)$ est vraie $\Longrightarrow$ $P(n+1)$ est vraie pour un rang donné $n$.<br><br>
         <u> Conclusion:</u><br>
        Comme  $P(n)$ est initialisée et héréditaire, elle est vraie <b><u>pour tout entier $n \geq 1$ </u></b>.
     
     
</div>

## III) Comportements des suites
### 1) Monotonie

<div class="alert alert-block alert-danger">
 On dit que $(U_n)_{n\in \mathbb{N}}$ est croissante $\iff$ $U_{n+1} \geq U_{n}$ pour tout n $\in \mathbb{N}$.  <br>
On dit que $(U_n)_{n\in \mathbb{N}}$ est décroissante $\iff$ $U_{n+1} \leq U_{n}$ pour tout n $\in \mathbb{N}$. <br>

On dit que $U_n$ est **monotone** à partir du rang $n_0$ lorsqu'elle est strictement croissante ou décroissante à partir de ce rang.

```python
def suite_croissante():
    U = 0
    L = [U]
    for i in range(4):
        U = U+100
        L.append(U)
    return L    
```

```python
suite_croissante()
```

```python
def suite_decroissante():
    U = 10
    L = [U]
    for i in range(4):
        U=U*0.2
        L.append(U)
    return L    
```

```python
suite_decroissante()
```

### 2) Suites majorées, minorées et bornées

<div class="alert alert-block alert-danger">
    Dire qu'une suite $(U_n)_{n \in \mathbb{N}}$ est : <br>
    - <b>majorée</b> signifie qu'il existe un réel $M$ tel que pour tout entier naturel $n$, $U_n \leq M$ <br>
    - <b>minorée</b> signifie qu'il existe un réel $m$ tel que pour tout entier naturel $n$, $U_n \geq m$ .<br>
    - <b>bornée</b>, signifie qu'elle est à la fois majorée et minorée.  
Il existe deux réels $m$ et $M$ tels que pour tout entier naturel $n$, $m \leq U_n \leq M$.<br>
    </div>

Pour illustrer nos propos, nous allons travailler sur le TP1 p 56 de votre ouvrage.

```python
def pop_cigognes(a,n):
    u = 0.3
    for i in range(n):
        u = a*u*(1-u)
    return u 
```

```python
round(pop_cigognes(1,15),3)
```

```python
import matplotlib.pyplot as plt

for a in range(1,5):
    x = [n for n in range(50)]
    y = [pop_cigognes(a,n) for n in range(50)]
    #plt.subplot(2,2,a)
    plt.scatter(x,y)
    plt.ylabel(a)
    plt.show()
```

### 3) Cas des suites géométriques

Une suite géométrique a pour relation de récurrence (ici $q$ réel positif):  
$$U_{n+1} = q \times U_n$$  
On peut étudier ses variations étudiant le signe de $U_{n+1}-U_n$.   
$U_{n+1}-U_n = q \times U_n - U_n $  
$U_{n+1}-U_n = U_n \times (q-1)$

Ce qui nous ramène à un étude de signe des deux facteurs $U_n$ et $q-1$.  
Si $U_n \gt 0$ et $q \gt 1$ alors la suite est strictement croissante pour tout entier naturel $n$.  
Si $U_n \gt 0$ et $q \lt 1$ alors la suite est strictement décroissante pour tout entier naturel $n$.  
Si $U_n \lt 0$ et $q \lt 1$ alors la suite est strictement croissante pour tout entier naturel $n$.  
Si $U_n \lt 0$ et $q \gt 1$ alors la suite est strictement décroissante pour tout entier naturel $n$.

**Exemple:**:  
1) Soit $(U_n)_{n \in \mathbb{N}}$ une suite géométrique avec $U_0 = 1$ et $q=\frac{1}{4}$ alors ...  
  
    

2)Soit $(U_n)_{n \in \mathbb{N}}$ une suite géométrique avec $U_0 = -5$ et $q=5$ alors ...  
  
    
    

**Attention:
Pour des valeurs de $q \lt 0$, la suite n'est plus monotone.**

```python
import matplotlib.pyplot as plt

n = [i for i in range(10)]
U = [(-2)**i for i in range(10)]
plt.scatter(n,U)
plt.show()
```

### 1) Limite de suites convergentes

Dire qu'une suite a pour limite $l$ pour $n$ tendant vers $+\infty$, signifie que l'intervalle ouvert contenant $l$, $[l-r;l+r]$, contient tous les termes de la suite à partir d'un certain rang.  
On dit alors que $(U_n)$ est **convergente**, sa limite étant $l$.  


La limite d'une suite convergente se note: $$\lim_{n \to +\infty}U_n=l$$

<div class="alert alert-block alert-danger">
    <b>Théorème:</b>  <br>
- Si $(U_n)$ est croissante et majorée, alors $(U_n)$ converge. <br> 
- Si $(U_n)$ est décroissante et minorée, alors $(U_n)$ converge.  <br>
    </div>

Appuyons nous sur les exercices 9 et 12 p 60-61 de votre ouvrage.

```python
  def seuil(p):
        n = 1
        while 1/n**3>=10**(-p):
            n=n+1
        return n, 1/n**3    
```

```python
seuil(4)
```

```python
from math import sqrt 
def rang_seuil():
    phi = (1+sqrt(5))/2
    n = 0
    u = 1
    while abs(u-phi)>= 10**(-7):
        u = sqrt(1+u)
        n=n+1
    return n    
```

```python
rang_seuil()
```

### 2) Divergence d'une suite

<div class="alert alert-block alert-info">
    <b>Démonstration: Une suite croissante non majorée est-elle toujours convergente? </b> <br>
Voici les trois assertions qui vont nous permettre de traduire cette question en langage mathématique. <br>   
 
> $1^e$ étape: Tentez de les reconnaître à partir du vocabulaire de la question.  <br>
 >  <br>
> (1)  Pour tout entier naturel $n$, il existe un réel A tel que $U_n \gt A$.   <br>
> (2)  Pour tout entier naturel $n$, $U_{n+1} \geq U_n$.  <br>
> (3)  Pour tout réel A, il exsite un entier naturel $p$ tel que pour tout entier $n \geq p$, $U_n\geq A$.<br>  

<br><br><br>

> $2^e$ étape: Faire un schéma représentatif sur un axe. <br>
>   <br>
>   <br>
>   <br>  
>  <br> 


Soit A un réel quelconque. On choisit un entier naturel $p$ tel que $U_p \geq A$, dont l'existence est assurée puisque $(U_n)_{n \in \mathbb{N}}$ n'est pas majorée. <br> 
Comme la suite $(U_n)$ est croissante, pour tout entier naturel $n \geq p$,   $U_{n+1} \geq U_{n} ... \geq U_p \gt A$ donc $U_n \gt A$.  <br>
On conclut que pour tout réel A, il existe un entier naturel $p$ tel que pour tout entier $n$ tel que $ n\geq p$, $U_n \geq A$. 
C'est-à-dire par définition de la limite:  
$$\lim_{n \to +\infty}U_n = +\infty $$

</div>

Une suite qui n'est pas convergente est **divergente**. 

Une suite qui diverge vers $+\infty$ se note, $\lim_\limits{n \to +\infty}U_n = +\infty$.  
Une suite qui diverge vers $-\infty$ se note, $\lim_\limits{n \to +\infty}U_n = -\infty$.  
Certaines suites divergentes n'ont pas de limite comme $U_n = (-1)^n$, définie pout tout $n$ naturel.

<div class="alert alert-block alert-danger">
    <b>Théorème:</b> <br>
Si $(U_n)$ est croissante et non majorée, alors $(U_n)$ diverge vers $+\infty$. <br> 
Si $(U_n)$ est décroissante et non minorée, alors $(U_n)$ diverge vers $-\infty$.<br>

```python
import matplotlib.pyplot as plt

def divergence():
    n= [i for i in range(20)]
    U = [(-1)**i for i in range(20)]
    plt.scatter(n,U)
    plt.show()
        
```

```python
divergence()
```

### 3) Propriétés admises sur les limites

Certaines suites divergent (ici k naturel non nul): 
$$\lim\limits_{n \to +\infty} \sqrt n = +\infty$$
$$\lim\limits_{n \to +\infty}  n^k = +\infty$$   
D'autres convergent:
$$\lim\limits_{n \to +\infty} \frac{1}{\sqrt n} = 0 $$  
$$\lim\limits_{n \to +\infty} \frac{1}{ n^k} = 0 $$

#### a) somme de limites

La limite d'une somme de suites est la somme de leur limite.   
**Exemple**:  
$\lim_\limits{n \to +\infty}\frac{1}{\sqrt n} = 0$ et $\lim_\limits{n \to +\infty} \sqrt n = +\infty$ alors $\lim_\limits{n \to +\infty}\frac{1}{\sqrt n} + \sqrt n = +\infty$.  Généralisable. 
*********************

<div class="alert alert-block alert-danger">
Mais il existe une forme dite indéterminée:
$\lim_\limits{n \to +\infty}U_n = -\infty$ et $\lim_\limits{n \to +\infty}V_n = +\infty$ . <br>  
Ici, aucune solution automatique n'est possible pour la limite de $U_n+V_n$. <br>
   Il faut transformer l'écriture pour déterminer sa limite.
</div>

#### b) produit de limites

La limite d'un produit de suites est le produit de leur limite.  
**Exemple:** (ici k est un entier naturel non nul)  
$\lim_\limits{n \to +\infty}{\sqrt n} = +\infty $ et $\lim_\limits{n \to +\infty}  n^k = +\infty$ alors $\lim_\limits{n \to +\infty}{\sqrt n} \times  n^k = +\infty$. Généralisable en appliquant la règle des signes.

<div class="alert alert-block alert-danger">
Mais il existe une forme dite indéterminée:
$\lim_\limits{n \to +\infty}{\sqrt n} = +\infty $ et $\lim_\limits{n \to +\infty}  \frac{1}{\sqrt n} = 0$.  <br>   
Ici, aucune solution automatique n'est possible pour la limite de $U_n \times V_n$. <br>
   Il faut transformer l'écriture pour déterminer sa limite.
</div>

#### c) quotient de limites

La limite d'un quotient de suites est le quotient de leur limite.  
**Exemple:** (ici k est un entier naturel non nul)  
$\lim_\limits{n \to +\infty}\frac{1}{\sqrt n}+4 = 4 $ et $\lim_\limits{n \to +\infty}  n^k = +\infty$ alors $\lim_\limits{n \to +\infty}\frac{\frac{1}{\sqrt n}+4}{n^k}= 0$. Généralisable .

<div class="alert alert-block alert-danger">
Mais il existe deux formes dites indéterminées:  <br>  
    
   -  $\lim_\limits{n \to +\infty}{\sqrt n} = +\infty $ et $\lim_\limits{n \to +\infty}  n^k = +\infty$.
Aucune solution automatique pour la limite de $\frac{U_n}{V_n}$. <br><br>
    
   -  $\lim_\limits{n \to +\infty}\frac{1}{\sqrt n} = 0 $ et $\lim_\limits{n \to +\infty} \frac{1}{ n^k} = 0 $. 
Aucune solution automatique pour la limite de $\frac{U_n}{V_n}$. <br><br> 
   Il faut transformer l'écriture pour déterminer sa limite.
</div>

## V) Théorème des comparaisons

<div class="alert alert-block alert-danger">
    Soient $(U_n)_{n \in \mathbb{N}}$ et $(V_n)_{n \in \mathbb{N}}$ deux suites telles qu'il existe deux entiers naturels $n$ et $p$ tels que pour $ n\geq p$, $(U_n)_{n \in \mathbb{N}} \leq (V_n)_{n \in \mathbb{N}}$.<br>

Si $\lim \limits_{n \to +\infty} U_n = +\infty$ alors $\lim \limits_{n \to +\infty} V_n = +\infty$  <br>
SI $\lim \limits_{n \to +\infty} V_n = -\infty$ alors $\lim \limits_{n \to +\infty} U_n = -\infty$  <br>
</div>

<div class="alert alert-block alert-info">
    <b>Démonstration: Divergence d'une suite minorée par une suite divergeant vers $+\infty$ .</b> <br><br>
Voici les trois assertions qui vont nous permettre de traduire cette question en langages mathématiques. <br>   
 
<b>$1^e$ étape:</b> Tentez de les reconnaître à partir du vocabulaire de la question.  <br> <br>
(1) $\lim\limits_{n \to +\infty} V_n = +\infty$  
(2)  Il existe un entier naturel q tel que pour tout entier naturel $n$, $n \geq p \Longrightarrow U_n \leq V_n$.  
(3)   Il existe un entier naturel $p$ tel que pour tout entier natruel $n$ $n \geq p$, $U_n \geq A$.

<b>$2^e$ étape:</b> Réalisation d'un schéma  <br> <br><br>

Soit A un réel quelconque. On choisit un entier naturel $p$ tel que pour tout entier natruel $n$ $n \geq p$, $U_n \geq A$ dont l'existence est assurée puisque $\lim\limits_{n \to + \infty} U_n = +\infty$.  <br>
On choisit un entier naturel q tel que pour tout entier naturel $n$, $n \geq p \Longrightarrow U_n \leq V_n$ dont l'existence est assurée par le fait que $(V_n)$ est minorée par $(U_n)$.<br>
En posant $r=max(p,q)$, on obtient pour tout entier naturel $n$, $n \geq r \Longrightarrow V_n \geq U_n \geq A$. On conclut que $\lim\limits_{n \to +\infty} V_n = +\infty$


</div>

<div class="alert alert-block alert-info">
    <b>Démonstration: Divergence de la fonction exponentielle vers $+\infty$ et $-\infty$ .</b> <br><br>
Nous allons nous appuyer sur le théorème des comparaisons en l'adaptant à notre démonstration.  
 
Soient $(U_n)_{n \in \mathbb{N}}$ et $(e^n)_{n \in \mathbb{N}}$ deux suites telles qu'il existe deux entiers naturels $n$ et $p$ tels que pour $ n\geq p$, $(U_n)_{n \in \mathbb{N}} \leq (e^n)_{n \in \mathbb{N}}$.

Si $\lim \limits_{n \to +\infty} U_n = +\infty$ alors $\lim \limits_{n \to +\infty} e^n = +\infty$ <br>
> Pour la limite en $+\infty$, il nous suffit de prendre une suite que l'on sait être divergente vers $+\infty$. (indice: passez aux suites géométriques)<br>
>  .

>  .     
>  .    
>  .
<br>
<br>
> Pour la limite en $-\infty$, il nous faut changer de variable car $n$  est un entier naturel, il ne peut tendre que vers $+\infty$. <br>
> On prendra donc $N=-n$, ainsi $\lim\limits_{N \to -\infty} e^N = \lim\limits_{n \to +\infty} e^{-n}= \lim\limits_{n \to +\infty}\frac{1}{e^n}$. <br> Ici aussi, on va s'appuyer sur les suites géométriques et le théorème des comparaisons.




## VI)  Théorème des gendarmes

<div class="alert alert-block alert-danger">
    Soient $p$ un entier naturel et $l$ un réel.<br>  
Soient $(U_n)$, $(V_n)$ et $(W_n)$ trois suites telles que pour tout $n \geq p$, $U_n \leq V_n \leq W_n$.<br>

Si $\lim\limits_{n \to +\infty}U_n = l$ et $\lim\limits_{n \to +\infty}W_n = l$ , alors $\lim\limits_{n \to +\infty}V_n = l$.<br>
